```python
const http = require('http')

const server = http.createServer((req, res) => {
    if (req.url === '/') {
        // Use return to stop the function execution here
        return res.end('Welcome to our homepage')
    }
    
    if (req.url === '/about') {
        return res.end('Here is the about page')
    }

    // This only runs if none of the above matched
    res.end(`
        <h1>Oops!</h1>
        <p>We can't seem to find the page you are looking for</p>
        <a href="/">back home</a>
    `)
})

server.listen(5000)
```

```python
//basic info project

const http = require('http');
const fs = require('fs');

const server = http.createServer((req, res) => {
    const url = req.url;
    const method = req.method;
    const filePath = 'users.json'; // Changed to .json for better practice

    if (url === '/') {
        let historyHtml = "<li>No data yet</li>";
        
        // 1. Read existing history
        if (fs.existsSync(filePath)) {
            try {
                const rawData = fs.readFileSync(filePath, 'utf8');
                const usersArray = JSON.parse(rawData);
                
                // 2. Build an HTML list of all users
                historyHtml = usersArray.map(user => 
                    `<li><strong>${user.username}</strong> (gender: ${user.gender}  Age: ${user.age})</li>`
                ).join('');
            } catch (err) {
                historyHtml = "<li>Error loading history</li>";
            }
        }

        res.setHeader('Content-Type', 'text/html');
        res.write('<html><body>');
        res.write('<h2>Submit Details</h2>');
        res.write('<form action="/submit-details" method="POST">');
        res.write('<input type="text" name="username" placeholder="Name" required><br><br>');
        res.write('<input type="text" name="gender" placeholder="Gender" required><br><br>');
        res.write('<input type="number"name="age"placeholder="Age" required><br><br>');
        res.write('<button type="submit">Submit</button>');
        res.write('</form>');
        
        res.write('<h3>Submission History:</h3>');
        res.write(`<ul>${historyHtml}</ul>`);
        res.write('</body></html>');
        return res.end();
    }

    if (url === "/submit-details" && method === "POST") {
        const body = [];
        req.on('data', (chunk) => body.push(chunk));

        return req.on('end', () => {
            const fullBody = Buffer.concat(body).toString();
            const params = new URLSearchParams(fullBody);
            const newUser = Object.fromEntries(params);

            // 3. Update the Array History
            let usersArray = [];
            if (fs.existsSync(filePath)) {
                const fileData = fs.readFileSync(filePath, 'utf8');
                usersArray = JSON.parse(fileData);
            }
            
            usersArray.push(newUser); // Add the new person to the list

            // 4. Save the updated list
            fs.writeFileSync(filePath, JSON.stringify(usersArray, null, 2));

            res.statusCode = 302;
            res.setHeader('Location', '/');
            return res.end();
        });
    }
});

server.listen(3000, () => console.log('Server running on http://localhost:3000'));
```
